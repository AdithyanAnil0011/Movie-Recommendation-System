```python
import pandas as pd
import numpy as np
import difflib    # even though there is some mistake in movie name (it will find the closest movie name)
from sklearn.feature_extraction.text import TfidfVectorizer  #for converting the textual data into numerical values
from sklearn.metrics.pairwise import cosine_similarity  #for finding the similarity scores(To find which all movies are related, an important feature for recommending similar movies)
```

```python
data=pd.read_csv('/content/moviesdata.csv')
```

```python
data.head()
```

```python
data.shape
```

```python
selected_features=['genres','keywords','overview','cast','director']
print(selected_features)
```

```python
#replace nullvalues with null string

for feature in selected_features:
  data[feature]=data[feature].fillna('')
```

```python
#combining the selected features

combined_features=(data['genres']+' '+data['keywords']+' '+' '+data['overview']+' '+data['cast']+' '+data['director'])
```

```python
print(combined_features)
```

```python
vectorizer=TfidfVectorizer()
```

```python
feature_vectors=vectorizer.fit_transform(combined_features)
```

```python
print(vectorizer.get_feature_names_out())
```

```python
print(feature_vectors)
```

```python
feature_names = vectorizer.get_feature_names_out()

print(feature_names[201])
print(feature_names[14378])
print(feature_names[4945])
print(feature_names[11503])
```

cosine similarity

```python
similarity=cosine_similarity(feature_vectors)
```

```python
print(similarity)
```

```python
print(similarity.shape)
```

```python
# getting the movie name from user

movie_name=input('Enter your favourite movie name: ')
```

```python
#creating a list of all movie names given in the dataset
list_of_all_titles=data['title'].tolist()  #data['title'] selects the title column(its a pandas series)
print(list_of_all_titles)  # tolist() converts the pandas series into a normal python list
```

```python
find_close_match=difflib.get_close_matches(movie_name,list_of_all_titles,n=10,cutoff=0.5)
print(find_close_match)
```

```python
close_match=find_close_match[0]
print(close_match)
```

```python
#finding the idx of the movie with title

index_of_the_movie=data[data.title==close_match]['index'].values[0]
print(index_of_the_movie)
```

```python
#getting the list of similar movies

similarity_score=list(enumerate(similarity[index_of_the_movie]))  #enumerate is used to have a loop in the list
print(similarity_score)
```

```python
len(similarity_score)
```

```python
#sorting the movies based on their similarity score

sorted_similar_movies=sorted(similarity_score,key=lambda x:x[1],reverse=True)
print(sorted_similar_movies)
```

```python
#print the name of similar movies based on the index

print('Movies suggested for you:\n')
i=1
for movie in sorted_similar_movies:
  index=movie[0]
  title_from_index=data[data.index==index]['title'].values[0]
  if(i < 10):
    print(i,'.',title_from_index)
    i+=1
```

MOVIE RECOMMENDATION SYSTEM


```python
  movie_name=input('Enter your favourite movie name: ')

  list_of_all_titles=data['title'].tolist()

  find_close_match=difflib.get_close_matches(movie_name,list_of_all_titles,n=10,cutoff=0.68)

  if len(find_close_match) == 0:
      print("Movie not found in dataset")
  else:

        close_match=find_close_match[0]

        index_of_the_movie=data[data.title==close_match]['index'].values[0]

        similarity_score=list(enumerate(similarity[index_of_the_movie]))

        sorted_similar_movies=sorted(similarity_score,key=lambda x:x[1],reverse=True)

        print('Movies suggested for you:\n')
        i=1
        for movie in sorted_similar_movies:
          index=movie[0]
          title_from_index=data[data.index==index]['title'].values[0]
          if(i < 10):
            print(i,'.',title_from_index)
            i+=1
```
